```python
import pandas as pd
df = pd.read_csv('vgsales.csv')
df.describe()
df.isnull().sum()
```

```python
df.dropna(subset=['Year'], inplace=True)
df.isnull().sum()

df.dropna(subset=['Publisher'], inplace=True)
df.isnull().sum()
```

```python
#Change platform name 2600 to Atari 2600
df.loc[df['Platform'] == '2600', 'Platform'] = 'Atari 2600'
```

```python
import plotly.express as px

fig = px.treemap(
    df, path=['Platform', 'Publisher', 'Genre'],
    values='Global_Sales',
    color='Year',
)

fig.update_layout(
    title="Game Sales Treemap",
    title_x=0.5,
    font=dict(size=14),
    legend_title="Game Genre",           # Legend title
    legend=dict(                         # Customize legend style
        x=1.05,                          # Position to the right of the plot
        y=1,                             # Position at the top
        borderwidth=1,                   # Add a border around the legend
        bgcolor="rgba(255,255,255,0.8)", # Transparent white background
        font=dict(size=12)               # Adjust legend font size
    )
)

fig.show()
```
