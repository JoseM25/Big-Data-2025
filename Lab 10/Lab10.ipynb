```python
import numpy as np
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler
```

## Círculos Concéntricos

### Cargar y crear dataset

```python
cc_file = files.upload()
```

```python
cc_dataset = pd.read_csv("circulos_concentricos.csv")
```

### Visualizar datos

```python
plt.scatter(cc_dataset['Feature1'], cc_dataset['Feature2'], s = 30)
plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Visualizacion de datos: Circulos Concentricos')
plt.grid(True)
plt.show()
```

### Escalar datos

```python
scaler = StandardScaler()
cc_data = scaler.fit_transform(cc_dataset[['Feature1', 'Feature2']].values)
```

### Clustering con KMeans

```python
cc_kmeans = KMeans(n_clusters = 2, random_state = 3456)
cc_kmeans.fit(cc_data)
```

```python
plt.scatter(cc_dataset['Feature1'], cc_dataset['Feature2'], c = cc_kmeans.labels_, s = 30)
```

### Clustering con DBSCAN

```python
cc_dbscan = DBSCAN(eps = 0.348, min_samples = 4)
cc_dbscan.fit(cc_data)
```

```python
plt.scatter(cc_dataset['Feature1'], cc_dataset['Feature2'], c = cc_dbscan.labels_, s = 30)
```

### Análisis de Resultados

Se puede observar como K-Means presenta dificultades a la hora de realizar el clustering debido a que está diseñado para agrupar linealmente, por lo que acaba dividiendo el conjunto de datos directamente por la mitad. Por otra parte, DBSCAN sí logra separar los datos de una mejor forma, pero no se consigue separar ambos anillos completamente con la mejor configuración alcanzada. Un número más grande para eps termina provocando que todo el conjunto se agrupe en un solo cluster, mientras que disminuirlo genera nuevos clusters pequeños que no parecen tener relevancia en este caso.

Esto puede deberse a que los clusters son capaces de crearse linealmente y existe una separación considerable entre los mismos, sin que existan datos que se separen mucho de alguno de estos.
