## Imports

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
from mlxtend.preprocessing import TransactionEncoder
import numpy as np
import pandas as pd
from google.colab import files
import time
```

```python
# Obtener dataset:
transactions = pd.read_csv('retail_data.csv')
transactions.head()
```

```python
# Eliminar columnas con valores nulos:
transactions = transactions.dropna()
len(transactions)
```

```python
transactions.drop(['Trans_ID'], axis=1, inplace=True)
transactions.head()
```

```python
transactions['product_list'] = (
    transactions[['Product1', 'Product2', 'Product3']]
    .apply(lambda row: row.dropna().tolist(), axis=1)
)

print(transactions[['product_list']].head())
```

## **2. T√©cnica para reglas de asociaci√≥n: FP-Growth trees**

```python
tiempo_fpgrowth_ini = time.perf_counter()
te = TransactionEncoder()
te_ary = te.fit_transform(transactions['product_list'].tolist())
encoded_transactions = pd.DataFrame(te_ary, columns=te.columns_, dtype=bool)

frequent_itemsets = fpgrowth(encoded_transactions, min_support=0.1, use_colnames=True)
association_rules_confidence = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.5)
tiempo_fpgrowth_fin = time.perf_counter() - tiempo_fpgrowth_ini

frequent_itemsets = (
      frequent_itemsets
      .sort_values(by='support', ascending=False)
      .reset_index(drop=True)
)

confidence_rules = (
      association_rules_confidence
      .sort_values(by='confidence', ascending=False)
      .reset_index(drop=True)
)

frequent_itemsets['itemset_disp'] = (
    frequent_itemsets['itemsets'].apply(lambda s: ', '.join(sorted(s)))
)
confidence_rules['antecedents'] = confidence_rules['antecedents'].apply(lambda s: ', '.join(sorted(s)))
confidence_rules['consequents'] = confidence_rules['consequents'].apply(lambda s: ', '.join(sorted(s)))

print("\n-------------------Frequent itemsets----------------\n")
print(frequent_itemsets[['itemset_disp', 'support']].to_string(index=False))

print("\n----------------Association rules---------------\n")
print(confidence_rules[['antecedents', 'consequents', 'support', 'confidence']].to_string(index=False))
```

```python
print("Tiempo ejecuci√≥n Algoritmo FPGrowth: ", tiempo_fpgrowth_fin, " segundos")
```

## **3. Selecci√≥n de la medida: Kulczynski y Conviction**

```python
association_rules_kulc = association_rules(frequent_itemsets, metric="kulczynski", min_threshold=0.6)

kulc_rules = (
    association_rules_kulc
    .sort_values(by='kulczynski', ascending=False)
    .reset_index(drop=True)
)


kulc_rules['antecedents'] = kulc_rules['antecedents'].apply(lambda s: ', '.join(sorted(s)))
kulc_rules['consequents'] = kulc_rules['consequents'].apply(lambda s: ', '.join(sorted(s)))

print("\n----------------Association rules (Kulczynski 0.6)---------------\n")
print(kulc_rules[['antecedents','consequents','support','kulczynski']].to_string(index=False))


association_rules_conv = association_rules(frequent_itemsets, metric="conviction", min_threshold=1.2)

conviction_rules = (
    association_rules_conv
    .sort_values(by='conviction', ascending=False)
    .reset_index(drop=True)
)

conviction_rules['antecedents'] = conviction_rules['antecedents'].apply(lambda s: ', '.join(sorted(s)))
conviction_rules['consequents'] = conviction_rules['consequents'].apply(lambda s: ', '.join(sorted(s)))

print("\n----------------Conviction rules (Conviction 1.2)---------------\n")
print(conviction_rules[['antecedents','consequents','support','conviction']].to_string(index=False))
```

## **4. Patrones Interesantes**

```python
kulc_rules_subset = (
    kulc_rules
    .loc[(kulc_rules['kulczynski'] > 0.8)]
    .sort_values(['kulczynski'],
                 ascending=[False])
)

conf_rules_subset_kulc = (
    confidence_rules
    .loc[(confidence_rules['kulczynski'] <= 0.5)]
    .sort_values(['kulczynski'],
                 ascending=[False])
)

conv_rules_subset = (
    conviction_rules
    .loc[(conviction_rules['conviction'] > 10)]
    .sort_values(['conviction'],
                 ascending=[False])
)

conf_rules_subset_conv = (
    confidence_rules
    .loc[(confidence_rules['conviction'] <= 1.0)]
    .sort_values(['conviction'],
                 ascending=[False])
)

print("\n----------------Association rules Kulc > 0.8)---------------\n")
print(kulc_rules_subset[['antecedents','consequents','support', 'confidence', 'kulczynski']].to_string(index=False))

print("\n----------------Association rules Kulc (Filtered)---------------\n")
print(conf_rules_subset_kulc[['antecedents','consequents','support', 'confidence', 'kulczynski']].to_string(index=False))

print("\n----------------Association rules Conv > 10)---------------\n")
print(conv_rules_subset[['antecedents','consequents','support', 'confidence', 'conviction']].to_string(index=False))

print("\n----------------Association rules Conv (Filtered)---------------\n")
print(conf_rules_subset_conv[['antecedents','consequents','support', 'confidence', 'conviction']].to_string(index=False))
```
